```python
main = reload(main)
all_params = []
all_llh = []
val = []
tr = []
all_x = []
n, data = main.read_data('sp500.mat')
data_train = data[:30]
data_valid = data[30:]
last_llh = 0
for i, (params, x_good) in enumerate(main.infer(data_train)):
    all_params += [params]
    train_llh = main.compute_llh(data_train, params)
    valid_llh = main.compute_llh(data_valid, params)

    all_llh += [[train_llh, valid_llh]]
    val += [valid_llh]
    tr += [train_llh]
    all_x += [x_good]
    if abs(train_llh - last_llh) < 1e-2:
        break
    last_llh = train_llh
```

```python
a = np.array(all_params)
a, b, c, d, = plt.plot(a[:, :])
#plt.yscale('log')
plt.savefig('params.png')
plt.xlabel('Iteration')
plt.ylabel('Parameter value')
plt.legend([a, b, c, d], ['a', 'b', 'c', 'd'])
plt.show()
```

```python
#a = np.array(zip(tr, val))
training, testing = plt.plot(tr, 'b', val, 'g')
plt.xlabel('Iteration')
plt.ylabel('Average loglikelihood')
plt.legend([training, testing], ['Training', 'Validation'], loc=4)
plt.savefig('likelihood_normalised.png')
plt.show()
```
